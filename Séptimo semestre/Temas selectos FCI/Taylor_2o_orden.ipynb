# Introducción

Vamos a buscar implementar el taylor para segundo orden cuando se tiene la ecuación:

$$\ddot x=f(x)$$

De aquí lo podemos separar en un sistema de dos ecuaciones de primer orden:
$$\dot x=v$$
$$\dot v=f(x)$$

Vamos a tener que hacer la integración en paralelo de los dos métodos, tomando entonces que vamos a tomar:

Sabemos que los coeficientes de Taylor de $v(t)$ van a estar dados por:
$$v_{n+1}=\frac{f_n(x)}{n+1}$$
Por lo cual:
$$x_{n+1}=\frac{v_n}{n+1}$$
Llegando finalmente a que:
$$x_{n+2}=\frac{f_n(x)}{n+2}$$

Teniendo que debemos de tener las condiciones iniciales:
$$x(0)=x_0,\ v(0)=v_0$$
De dónde se encuentran los primeros valores de la serie:
$$x_0=x(0),\ x_1=v(0)$$

Vamos a buscar hacer la integración del oscilador armónico, el cual nos da la ecuación:
$$\ddot x=-\omega x$$
De dónde $f(x)=-\omega x$

# Integración

```julia
const orderMax=18
const epsilon=1e-30
```

```julia
x0=1.0
v0=0.0
C0=[x0,v0]
```

Vamos a tomar $\omega=1$

```julia
f(x,ω=1)=-ω*x
```

```julia
function NextTaylor(f::Function,C0)
    Q=[C0[1]]
    V=[C0[2]]
    for ord in 1:orderMax
        Q0=Taylor(Q)
        c1=f(Q0).coeffs[ord]/ord
        c2=V[ord]/ord
        push!(V,c1)
        push!(Q,c2)
    end
    Q,V
end
```

```julia
Q,V=NextTaylor(f,C0)
```

```julia
function TimeStep(Q,V)
    #Cuando implementemos en n cuerpos, es mejor si vamos sacando los
    #mínimos poco a poco (sea este mínimo que sacamos para cada coord.
    #de cada cuerpo, luego sacar el mínimo para cada cuerpo y
    #hasta al final sacar el #gran mínimo)
    A=Float64[]
    pot=inv(orderMax)
    if V[end]==0
    else
        push!(A,(epsilon/abs(V[end]))^pot)
    end
    if Q[end]==0
    else
        push!(A,(epsilon/abs(Q[end]))^pot)
    end
    pot=inv(orderMax)-1
    if V[end-1]==0
    else
        push!(A,(epsilon/abs(V[end-1]))^pot)
    end
    if Q[end-1]==0
    else
        push!(A,(epsilon/abs(Q[end-1]))^pot)
    end
    minimum(A)
end     
```

```julia
h=TimeStep(Q,V)
```

```julia
function EvalHorner(Q,V,h)
    m=length(Q)
    n=length(V)
    sumaQ=0.0
    sumaV=0.0
    for i in m-1:-1:1
        sumaQ=Q[i]+h*sumaQ
    end
    for i in n-1:-1:1
        sumaV=V[i]+h*sumaV
    end
    sumaQ,sumaV
end
```

```julia
EvalHorner(Q,V,h)
```

```julia
function TaylorStepper(f::Function,C0,tiempo)
    q,v=NextTaylor(f,C0)
    h=TimeStep(q,v)
    c1,c2=EvalHorner(q,v,h)
    tiempo+=h
    return c1,c2,tiempo
end
```

```julia
x0=1.0
v0=0.0
tiempo=0
X=Float64[x0]
T=Float64[tiempo]
while tiempo<300.0
    x0,v0,tiempo=TaylorStepper(f,[x0,v0],tiempo)
    println(tiempo,"\t",x0,"\t")
    push!(X,x0)
    push!(T,tiempo)
end
```
