```julia
function pasos(T)
    dT=Float64[]
    for i in 1:(length(T)-1)
        push!(dT,T[i+1]-T[i])
    end
    dT
end
```

```julia
dT=pasos(T);
```

```julia
maximum(dT)
```

```julia
minimum(dT)
```

```julia
function estandar(dE)
    final=length(dE)
    inicial=ceil(final/10)
    Std=Float64[]
    for i in inicial:final
        push!(Std,std(dE[1:i]))
    end
    [inicial:final],Std
end
```

```julia
j,Std=estandar(dE);
```

```julia
plot(j,Std)
```
