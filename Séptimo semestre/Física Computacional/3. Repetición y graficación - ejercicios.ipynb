###Tiro parabólico
####(6) - Haz una función que calcula la trayectoria de una partícula en tiro parabólico con una velocidad inicial dada

```julia
function tirazo(θ,v)
    seno=sin(θ)
    coseno=cos(θ)
    tangente=tan(θ)
    g=9.81
    tmax=2v*seno/g
    equis(t)=v*coseno*t
    x=[0:0.01:equis(tmax)]
    y=[(-0.5)*g*((z/(v*coseno))^2)+tangente*z for z in x]
    T=[z/(v*coseno) for z in x]
    plot(x,y,label=string(L"$\theta=\ $",round(θ/pi,2),"π"))
    xlabel(L"$x$")
    ylabel(L"$y$")
    title("Tiro parabólico")
    legend(loc=0, fontsize=10)
end
```

```julia
tirazo(pi/4,10)
```

```julia
function compara(paso,velocidad)
    angulos=[paso:paso:(pi/2-0.00000000001)]
for a in angulos
        tirazo(a,velocidad)
end
end
```
