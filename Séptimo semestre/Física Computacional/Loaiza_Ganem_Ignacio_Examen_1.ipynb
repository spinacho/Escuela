#### (i) Crea una función `fact` que utiliza un bucle para calcular el factorial de un entero, $n! = 1 \times 2 \times \cdots \times n$.   <p>(ii) Calcula el $10!$ y chécalo con el valor verdadero `factorial(10)`.   <p>(iii) ¿Cómo puedes hacer para que tu función funcione con números enteros de tamaño arbitrariamente grandes?  <p>(iv) Calcula $100!$

####(i)
Vamos a crear la función factorial utilizando un bucle **for**:

```julia
function fact(n::Integer)
    if n==0
        1
    end
    i=1
    for j in 1:n
        i=i*j
    end
    i
end
```

####(ii)
El valor que da $10!$ utilizando la función ya definida en Julia (sea el valor "verdadero") es:

```julia
factorial(10)
```

Y el valor obtenido utilizando la función definida en (i) es:

```julia
fact(10)
```

Notamos que es el mismo valor, lo cual quiere decir que la función está, en principio, bien implementada. :)

####(iii)
Vamos a agregar la función **big** dentro de la implementación de la función para poder hacer el cálculo con números arbitrariamente grandes:

```julia
function fact_big(n::Integer)
    if n==0
        1
    end
    if n<15
        return fact(n)
    end
    i=big(1)
    for j in 1:n
            i=i*big(j)
    end
    i
end
```

*nota*: Está implementación también se podría hacer nada más volviendo a $n$ un **big** desde el principio, o introduciendo el valor big(n) como argumento de la función orignial **fact(big(n))**, pero de esta forma se optimiza el tiempo de procesamiento de la función para valores pequeños, ya que no tiene que hacer la conversión a **big**.

```julia
fact(30)
```

```julia
fact_big(30)
```

Notamos como, por ejemplo, para $n=30$, esta implementación **big** de la función arroja un valor "coherente", mientras que la implementación normal arroja un resultado negativo (i.e. incoherente)

####(iv)
Vamos a calcular entonces a $100!$:

```julia
fact_big(100)
```

Notamos que es el mismo valor que se obtiene para la función factorial de Julia:

```julia
factorial(big(100))
```

#### (i) Crea una función `exponencial` que calcula el exponencial $\exp(x)$ de un número real $x$, usando un bucle  para calcular la serie de Taylor  $$\exp(x) = 1 + x + \textstyle \frac{1}{2!} x^2 + \frac{1}{3!} x^3 + \cdots.$$  Asegúrate de hacer el cálculo usando precisión arbitraria.  <p>El bucle no debe tener un número de términos fijo, sino terminar hasta que el tamaño del término nuevo que se va agregando sea menor que una tolerancia que tú especifiques.  <p>(ii) Utiliza tu función para calcular los exponenciales para los números $1$ hasta $10$ y calcula la diferencia entre el valor que obtienes y el valor verdadero.

####(i)

```julia
function mi_exp(x,ϵ=1e-20)
    X=Float64[]
    x=big(x)
    n=0
    while (x^n)/factorial(big(n))>ϵ
        push!(X,(x^n)/(factorial(big(n))))
        n=n+1
    end
    ll=length(X)
    return sum([X[i] for i in ll:-1:1])
end
```

*nota*: Hicimos que se sume desde el último hasta el primer término en vez de la forma "tradicional" ya que de esta forma el error de la computadora en las sumas es mucho menor.

####(ii)
Primero vamos a hacer una función "diferenciadora_exp" que obtenga el valor absoluto de la diferencia entre el valor de mi función con el valor de la función implementada en Julia:

```julia
diferenciadora_exp(x,ϵ=1e-20)=abs(mi_exp(x,ϵ)-exp(x))
```

Ahora vamos a hacer un arreglo de números que sean las diferencias entre el valor calculado y el valor teórico:

```julia
A=[diferenciadora_exp(i) for i in 1:10]
```

Notamos que el error más grande que dió la función fué de $2.27374*10^{-13}$. Podemos entonces concluir que la función implementada tiene una muy buena exactitud.

Vamos a graficar el error:

```julia
using PyPlot;
```

```julia
plot([1:10],A,"ro")
xlabel(L"$x$")
ylabel(L"$Abs(\Delta(e^x))$")
title("Valor absoluto de la diferencia");
```

#### <p>(i) Escribe una función que lleva a cabo el método de Euler.  <p>(ii) Utilízalo para resolver la ecuación diferencial ordinaria  $$\dot{x} = x(1-x)$$  que describe la dinámica de una población.  <p>(iii) Dibuja la solución para distintas condiciones iniciales positivas. Interpreta el resultado en términos de poblaciones.

####(i)

Sabemos que, para sacar la iteración $n+1$ en el método de Euler, se tiene que:
$$x_{n+1}=x_n+h*f(x)$$
Vamos a implementar está función.

```julia
function Euler(f::Function,x0,final=1000,h=1e-3)
    X=float([x0])
    for i in 1:final
        push!(X,X[i]+h*f(X[i]))
    end
    X
end
```

####(ii)
Ahora vamos a definir la función $ff(x)$ tal que $ff(x)=\dot x$:

```julia
ff(x)=x*(1-x)
```

Vamos a hacer una función que grafique la curva obtenida para $x_0=20$:

```julia
function graficadora(f::Function,x0,final=1000,h=1e-3)
    A=Euler(ff,x0,final,h)
    T=[0:final]
    plot(T,A,label="x(0)=$x0")
    xlabel(L"$n$")
    ylabel(L"$x_n$")
    title("Evolución del sistema")
    legend(loc=0, fontsize=10)
end
```

```julia
graficadora(ff,20);
```

####(iii)
Vamos a dibujar el sistema para varias condiciones iniciales positivas:

```julia
for x0 in 1:2:19
    graficadora(ff,x0)
end
```

Ahora para valores más grandes:

```julia
for x0 in 20:10:100
    graficadora(ff,x0)
end
```

Y ahora para valores entre $0$ y $1$:

```julia
for x0 in 0.1:0.1:1
    graficadora(ff,x0,10000)
end
```

Vamos a hacer ahora una gráfica interactiva:

```julia
using Interact;
```

```julia
fig = figure()
@manipulate for x0 in 1:100
    withfig(fig) do
        graficadora(ff,x0,4000)
        plot([1:4000],ones(4000),"r--",label="Asíntota x=1")
        legend()
        axis([0,4000,0,x0+1])
    end
end
```

Ya que tener sólo un individuo no tiene sentido, vamos a darle sentido a este sistema pensando que cada $x$ representa una colonia de alguna población de individuos. Podemos ver cómo, según va pasando el tiempo, el valor de $x$ va tendiendo a 1. Esto quiere decir que, sin importar cuantas colonias se tengan al principio, el número de estas va tendiendo a uno. En términos de poblaciones esto se podría deber a que en este sistema sólo hay comida suficiente para alimentar a una colonia, por lo cual sin importar cuantas haya al principio, siempre se tiende a una sola colonia. En el caso en el que se comienza con menos de una colonia, ya que hay más comida de la necesaria, la población va aumentando ya que hay comida de sobra y se están reproduciendo, tendiendo asintóticamente a una colonia.
