# ¿Qué queremos hacer?

Queremos resolver problemas de física, expresados como modelos matemáticos.

Antes: Tenemos que entender cómo "hacer matemáticas" en la computadora.

- Números: enteros, racionales, reales, complejos

- Funciones: una caja que coma algo y escupe otra cosa

Vamos a hacer aproximaciones y algoritmos.

Un *algoritmo* es una receta computacional.

- con series de Taylor
- con polinomios
- con Fourier
- métodos (seudo-)espectrales con polinomios de Chebyshev
