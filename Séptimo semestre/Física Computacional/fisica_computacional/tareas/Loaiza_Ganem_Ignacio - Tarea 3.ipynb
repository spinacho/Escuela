```julia
function derivada(f,x, h=1,i=0)
    ϵ=1e-5
    x=big(x)
    der(h)=(f(x+h)-f(x-h))/2h
    while abs(der(h)-der(10h))>ϵ
        h=h/10
        i=i+1
    end
    der(h)
end
```

```julia
function Newton(f,x,i=0)
    ϵ=1e-5
    x=big(x)
    while abs(f(x))>ϵ && i<50
        x=x-(f(x)/derivada(f,x))
        i=i+1
    end
    x
end
```

Notamos que, al cabo de 100 iteraciones, el método se detiene y arroja el valor en el que estaba.

```julia
function matriz_Newton(f,a,paso=0.1)
    I=[-a:paso:a]
    b=length(I)
    M=complex(zeros(b,b))
    i=j=1
    c=d=-a
    while -a<=c<=a && -a<=d<=a
        z=c+im*d
        M[j,i]=Newton(f,z)
        i=i+1
        c=c+paso
        if i+j>1000
            return println("Murió")
        end
        if c>=a
            c=-a
            i=1
            j=j+1
            d=d+paso
        end
        if d>=a
            return M
        end
    end
end
```

####(a)
Tomando las ecuaciones:
$$\ddot x(t)=0$$
$$\ddot y(t)=-g$$
Se llega analíticamente a que:
$$x(t)=v_0\cos(\theta)t+x_o$$
$$y(t)=-\frac{1}{2}gt^2+v_o\sin(\theta)t+y_o$$

Para resolver estas ecuaciones, vamos a utilizar el método de Euler, en el cual la iteración $n+1$ está dada por:
$$x_{n+1}=x_n+h\dot x_n$$
$$y_{n+1}=y_n+k\dot y_n$$
Con $h$ y $k$ *muy* pequeñas.

Vamos entonces a utilizar que:
$$\ddot x(t)=0$$
$$\ddot y(t)=-g$$
Vamos entonces a evaluar a $\ddot x$ y a $\ddot y$

Implementemos el método de Euler:

```julia
function Euler(xdot,x0,t0,tfinal,h=1e-5)
    T=[t0:h:tfinal]
    M=float64(zeros(T))
    i=1
    l=length(M)
    x=x0
    t=t0
    M[1]=x0
    while i<l
        M[i+1]=h*xdot(x,t)+M[i]
        t=t+h
        i=i+1
        end
    M
end
```

```julia
function Euler_2do_Orden(xddot,xdot0,x0,t0,tfinal,h=1e-5)
    T=[t0:h:tfinal]
    XDOT=float64(zeros(T))
    X=XDOT
    i=1
    l=length(X)
    while i<=l
        XDOT[i]=xdot0
        X[i]=x0
        xdot0=h*xddot(xdot0,t0)+xdot0
        x0=h*xdot0+x0
        t0=t0+h
        i=i+1
    end
    X
end
```

```julia
function Euler_2do_Orden_2_Variables(xddot,yddot,xdot0,x0,ydot0,y0,t0,tfinal,h=1e-5,γ=1,titulo="Tiro parabólico")
    T=[t0:h:tfinal]
    XDOT=float64([xdot0])
    X=float64([x0])
    YDOT=float64([ydot0])
    Y=float64([y0])
    i=1
    l=length(T)
    while i<l
        push!(XDOT,h*xddot(XDOT[i],X[i],YDOT[i],Y[i],T[i],γ)+XDOT[i])
        push!(YDOT,YDOT[i]+h*yddot(XDOT[i],X[i],YDOT[i],Y[i],T[i],γ))
        push!(X,X[i]+h*XDOT[i])
        push!(Y,Y[i]+h*YDOT[i])
        i=i+1
    end
    plot(X,Y, label="$titulo")
    xlabel("X")
    ylabel("Y")
    title("$titulo")
    legend(loc=0, fontsize=14)
end
```

```julia
xddot(xdot,x,ydot,y,t,γ)=0.0
yddot(xdot,x,ydot,y,t,γ)=-9.81
```

Ahora vamos a hacer los conjuntos de valores teóricos:

```julia
function parabolico(xdot0,x0,ydot0,y0,t0,tfinal,h=1e-5)
    T=[t0:h:tfinal]
    X=float64([T])
    Y=float64([T])
    i=1
    l=length(T)
    for i in [1:l]
        X[i]=xdot0*T[i]+x0
        Y[i]=-0.5*9.81*((T[i])^2)+ydot0*T[i]+y0
    end
    plot(X,Y, label="Tiro parabólico analítico")
    xlabel("X")
    ylabel("Y")
    title("Tiro parabólico analítico")
    legend(loc=0, fontsize=14)
    
end
```

```julia
f = figure()
@manipulate for vx0 in 0:1:100, x0 in 0:1:50, vy0 in 0:1:100, y0 in -20:1:20
    withfig(f) do
        Euler_2do_Orden_2_Variables(xddot,yddot,vx0,x0,vy0,y0,0,10,1e-2)
        parabolico(vx0,x0,vy0,y0,0,10,1e-2)
    end
end
```

Esta función 'Interact' está un poco lenta, pero se puede observar como se obtuvo la misma curva para el modelo analítico que para el computacional.

Vamos a agregar la función con fricción proporcional a la velocidad. Las EDOs se ven entonces como:
$$\ddot r(t)=-g-\gamma v$$
Siendo esta una relación vectorial, se reduce a:
$$\ddot x(t)=-\gamma \dot x(t)$$
$$\ddot y(t)=-g-\gamma \dot y(t)$$
Colocamos estas nuevas funciones en el método de Euler y vamos a ver que se obtiene:

```julia
xddotfric(xdot,x,ydot,y,t,γ)=-γ*xdot
yddotfric(xdot,x,ydot,y,t,γ)=-9.81-γ*ydot
```

Se puede ver en la comparación cómo el tiro con fricción claramente avanza menos que el tiro parabólico analítico (sin fricción), además de que, al igualar a $\gamma=0$, se obtiene que el tiro parabólico con fricción se vuelve idéntico al tiro parabólico sin fricción.

```julia
function Euler_2do_Orden_2_Variables_parable(xddot,yddot,xdot0,x0,ydot0,y0,t0,h=1e-5,γ=1)
    T=float64([t0])
    XDOT=float64([xdot0])
    X=float64([x0])
    YDOT=float64([ydot0])
    Y=float64([y0])
    i=1
    while YDOT[i]+h*yddot(XDOT[i],X[i],YDOT[i],Y[i],T[i],γ)>=0
        push!(T,T[i]+h)    
        push!(XDOT,h*xddot(XDOT[i],X[i],YDOT[i],Y[i],T[i],γ)+XDOT[i])
        push!(YDOT,YDOT[i]+h*yddot(XDOT[i],X[i],YDOT[i],Y[i],T[i],γ))
        push!(X,X[i]+h*XDOT[i])
        push!(Y,Y[i]+h*YDOT[i])
        i=i+1
    end
    alcance=last(X)
    tiempo=last(T)
    [tiempo,alcance]
end
```

```julia
Euler_2do_Orden_2_Variables_parable(xddotfric2,yddotfric2,500,50,500,50,0,1e-3,1)
```

Esta función nos regresa un arreglo de dos entradas: la primera entrada es el tiempo, y la segunda es el alcance.

####(b)
Vamos a crear una función que grafique los tiempos y los alcances máximos en función del ángulo inicial.

```julia
function graficadora(xddot,yddot,θi,θf,pasoangulo,v0,x0,y0,h=1e-3,γ=1,titulo="Tiro parabólico")
    THETAS=float64([pasoangulo:pasoangulo:θf])
    xdot0=v0*cos(θi)
    ydot0=v0*sin(θi)
    A=Euler_2do_Orden_2_Variables_parable(xddot,yddot,xdot0,x0,ydot0,y0,0,h,γ)
    TIEMPOS=[first(A)]
    ALCANCES=[last(A)]
        for θ in THETAS
        xdot0=v0*cos(θ)
        ydot0=v0*sin(θ)
        A=Euler_2do_Orden_2_Variables_parable(xddot,yddot,xdot0,x0,ydot0,y0,0,h,γ)
        tiempo=first(A)
        alcance=last(A)
        push!(TIEMPOS,tiempo)
        push!(ALCANCES,alcance)
    end
    THETAS=append!([0.0],THETAS)
    plot(THETAS,TIEMPOS,label="Tiempo de vuelo")
    plot(THETAS,ALCANCES,label="Alcance máximo")
    xlabel(L"\theta")
    ylabel(L"Tiempo de vuelo y alcance maximo")
    title("$titulo")
    legend(loc=0)
end
```

Vamos a recordar la función del oscilador armónico:
$$\ddot x(t)=-\omega^2 x(t)$$

```julia
function Euler_2do_Orden_1_Variable_parable(xddot,xdot0,x0,t0,tfinal,h=1e-5)
    T=[t0:h:tfinal]
    XDOT=float64([xdot0])
    X=float64([x0])
    i=1
    l=length(T)
    while i<l
        push!(XDOT,h*xddot(XDOT[i],X[i],T[i])+XDOT[i])
        push!(X,X[i]+h*XDOT[i])
        i=i+1
    end
    T,X,XDOT
end
```

```julia
xddotosc(xdot,x,t,ω=1)=(-ω^2)*x
```

```julia
T,X,XDOT=Euler_2do_Orden_1_Variable_parable(xddotosc,1,0,0,100,0.1);
```

```julia
plot(T,X, label="Oscilador armónico")
    xlabel("t")
    ylabel("x(t)")
title("Oscilador armónico: posiciones en función del tiempo")
    legend(loc=0, fontsize=14)
```

```julia
plot(XDOT,X, label="Oscilador armónico")
xlabel(L"$\dot x(t)$")
    ylabel("x(t)")
title("Oscilador armónico: posiciones en función de la velocidad")
    legend(loc=0, fontsize=14)
```

Notamos que el método de Euler no funciona muy bien en este caso, ya que existe un error que va creciendo cada vez más. Como primer intento de "arreglar" este error, se podría correr a la función con un paso mucho más pequeño como se va a ver en el siguiente cuadro. Sin embargo, este método sólamente va a aumentar el tiempo en el que sucede el error, sin eliminarlo. Como segunda aproximación a arreglar este error se podrían utilizar algoritmos de otro estilo, como intentar un acercamiento en base a polinomios de Taylor de ordenm "grande" (cómo de orden 18). Este método se basa en hacer una aproximación a la función como lo hace Euler pero condos grandes diferencias:
- Primero, este método toma muchos más términos del polinomio de Taylor, teniendo entonces una exactitud más grande.
- Segundo, y más importante, el paso de iteración de este método varía en función a los términos de la serie de Taylor que se está utilizando: utilizando los términos de la serie de Taylor, se puede saber que tan rápido va a converger la función. Se le pone entonces una tolerancia extremadamente pequeña (más pequeña que la diferencia entre los puntos flotantes de la computadora). Después, se calcula el siguiente paso en la iteración corriendo con el tiempo necesario para obtener al siguiente paso dentro de la tolerancia. Se vuelven entonces a calcular las componentes de la serie de Taylor, y se repite el proceso hasta llegar al tiempo deseado. Este método da entonces una serie de errores extremadamente pequeños y aleatorios, teniendo que, en vez de irse sumando los errores y volviéndose cada ves más grandes, estos se comportan como un movimiento Browniano.
