La Lagrangiana de un sistema se define cómo: (en coordenadas cartesianas)
$$L(x,y,z,v_x,v_y,v_z,t) = T-U =\frac{m}{2}(v_x^2+v_y^2+v_z^2) - V(x,y,z)$$
Sea:
$$L(x,y,z,\dot x,\dot y,\dot z,t) = \frac{m}{2}(\dot x^2+\dot y^2+\dot z^2) - V(x,y,z)$$
####$\mathcal{(i)}$
La Lagrangiana del sistema en coordenadas cartesianas es entonces (dónde $x^j_i$ es la coordenada $j-esima$ del péndulo $i$) (con $j\in {1,2};\ i\in {1,2})$:
<p> 
<p> 
$$L(x^j_i,\dot x^j_i,t)=\frac{1}{2}\sum_{i,j}m_j*(\dot x^j_i)^2-g(m_1x^1_2+m_2x^2_2)$$

```julia
t=sym"t"
x1=SymFunction("x_1")(t)
x2=SymFunction("x_2")(t)
y1=SymFunction("y_1")(t)
y2=SymFunction("y_2")(t)
m1,m2,l1,l2=sym"m_1,m_2,l_1,l_2"
g=sym"g";
```

```julia
T=SymFunction("T")(t)
U=SymFunction("U")(t)
L=SymFunction("L")(t);
```

```julia
vx1=diff(x1,t)
vx2=diff(x2,t)
vy1=diff(y1,t)
vy2=diff(y2,t);
```

```julia
T=(m1/2*(vx1^2+vy1^2)+m2/2*(vx2^2+vy2^2))
U=g*(m1*y1+m2*y2)
L=T-U
```

Ahora vamos a definir con las coordenadas "adecuadas": ($\theta_1 \ y \ \theta_2$).

```julia
θ1=SymFunction("theta_1")(t)
θ2=SymFunction("theta_2")(t)
x1=l1*sin(θ1)
x2=x1+l2*sin(θ2)
y1=-l1*cos(θ1)
y2=y1-l2*cos(θ2)
vx1=diff(x1,t)
vx2=diff(x2,t)
vy1=diff(y1,t)
vy2=diff(y2,t)
dθ1=SymFunction("\\dot{\\theta}_1")(t)
dθ2=SymFunction("\\dot{\\theta}_2")(t)
ddθ1=SymFunction("\\ddot{\\theta}_1")(t)
ddθ2=SymFunction("\\ddot{\\theta}_2")(t)
T=(m1/2*(vx1^2+vy1^2)+m2/2*(vx2^2+vy2^2))
U=g*(m1*y1+m2*y2)
L=T-U;
```

La función Lagrangiana en las coordenadas generalizadas es entonces:

```julia
L;
```

```julia
L=subs(L,diff(θ1,t),dθ1)
L=subs(L,diff(θ2,t),dθ2)
L=simplify(L)
```

####$\mathcal{(ii)}$
Las ecuaciones de Euler-Lagrange son:
$$\frac{d}{dt}\frac{\partial L}{\partial \dot q_i}-\frac{\partial L}{\partial q_i}=0$$
Con las coordenadas generalizadas $\theta_1$ y $\theta_2$. Se tienen entonces las ecuaciones siguientes:

```julia
L_dtheta1=simplify(diff(L,dθ1))
L_dtheta2=simplify(diff(L,dθ2))
L_theta1=simplify(diff(L,θ1))
L_theta2=simplify(diff(L,θ2))
Lt_1=simplify(diff(L_dtheta1,t))
Lt_2=simplify(diff(L_dtheta2,t));
```

```julia
Lt_1=subs(Lt_1,diff(θ1,t),dθ1)
Lt_1=subs(Lt_1,diff(dθ1,t),ddθ1)
Lt_1=subs(Lt_1,diff(θ2,t),dθ2)
Lt_1=subs(Lt_1,diff(dθ2,t),ddθ2)
Lt_2=subs(Lt_2,diff(θ1,t),dθ1)
Lt_2=subs(Lt_2,diff(dθ1,t),ddθ1)
Lt_2=subs(Lt_2,diff(θ2,t),dθ2)
Lt_2=subs(Lt_2,diff(dθ2,t),ddθ2)
L_theta1=subs(L_theta1,diff(θ1,t),dθ1)
L_theta1=subs(L_theta1,diff(dθ1,t),ddθ1)
L_theta1=subs(L_theta1,diff(θ2,t),dθ2)
L_theta1=subs(L_theta1,diff(dθ2,t),ddθ2)
L_theta2=subs(L_theta2,diff(θ1,t),dθ1)
L_theta2=subs(L_theta2,diff(dθ1,t),ddθ1)
L_theta2=subs(L_theta2,diff(θ2,t),dθ2)
L_theta2=subs(L_theta2,diff(dθ2,t),ddθ2);
```

```julia
eq1=Lt_1-L_theta1
eq2=Lt_2-L_theta2;
```

```julia
acel1,acel2=solve([eq1,eq2],[ddθ1,ddθ2]);
```

```julia
acel1=acel1[2]
acel2=acel2[2];
```

```julia
acel1=simplify(acel1)
acel2=simplify(acel2);
```

####$\mathcal{(iii)}$

Vamos a implementar el método de Euler-Cromer:

```julia
function euler_cromer(q0,v0,t0,tf,h,f_q,f_v,C)
    q = Vector{Float64}[]
    v = Vector{Float64}[]
    t = Float64[]
    a1=subs(acel1,m1,C[1])
    a1=subs(a1,m2,C[2])
    a1=subs(a1,l1,C[3])
    a1=subs(a1,l2,C[4])
    a1=subs(a1,g,C[5])
    a2=subs(acel2,m1,C[1])
    a2=subs(a2,m2,C[2])
    a2=subs(a2,l1,C[3])
    a2=subs(a2,l2,C[4])
    a2=subs(a2,g,C[5])
    for ti in t0:h:tf
        v_nueva = v0 + h*f_v(ti,q0,v0,C,a1,a2)
        q_nueva = q0 + h*f_q(ti,q0,v_nueva)
        q0 = q_nueva
        v0 = v_nueva
        push!(t,ti)
        push!(q,[q0])
        push!(v,[v0])
        
    end
    t,hcat(q...)',hcat(v...)'
end
```

Ahora hay que definir las funciones. Vamos a tomar la notación siguiente:
$$Q[1]=\theta_1,\ Q[2]=\theta_2,\ V[1]=\dot\theta_1,\ V[2]=\dot\theta_2$$
<p>
$$f_q(t,q,v)=\dot{\vec\theta}=[V[1],V[2]]$$
<p>
$$f_v(t,q,v)=\ddot{\vec\theta}(t,\vec\theta,\dot{\vec\theta})$$
Y finalmente:
$$C=[m_1,m_2,l_1,l_2,g]$$

```julia
fq(t,Q,V)=[V[1],V[2]]
```

```julia
function fv(ti,Q,V,C,a1,a2)
    v1=subs(a1,θ1,Q[1])
    v1=subs(v1,θ2,Q[2])
    v1=subs(v1,dθ1,V[1])
    v1=subs(v1,dθ2,V[2])
    v1=subs(v1,t,ti)
    v2=subs(a2,θ1,Q[1])
    v2=subs(v2,θ2,Q[2])
    v2=subs(v2,dθ1,V[1])
    v2=subs(v2,dθ2,V[2])
    v2=subs(v2,t,ti)
    [v1,v2]
end;
```

Cabe notar que, para utilizar esta función, ya hay que haber remplazado los valores de las constantes, lo cual se puso en la implementación de Euler-Cromer para evitar tener que hacer el remplazo en cada iteración.

```julia
t=Sym("t")
m,k=symbols("m,k", positive=true)
x1=Sym("x_1")
x2=Sym("x_2")
x3=Sym("x_3")
x4=Sym("x_4")
X=[x1,x2,x3,x4]
v1=Sym("\\dot{x_1}")
v2=Sym("\\dot{x_2}")
v3=Sym("\\dot{x_3}")
v4=Sym("\\dot{x_4}")
V=[v1,v2,v3,v4]
a1=Sym("\\ddot{x_1}")
a2=Sym("\\ddot{x_2}")
a3=Sym("\\ddot{x_3}")
a4=Sym("\\ddot{x_4}")
A=[a1,a2,a3,a4]
w=symbols("omega", positive=true);
```

Ecuaciones de Newton:
$$m\ddot x_1(t)=-kx_1-k(x_1-x_2)$$
$$m\ddot x_2(t)=-k(x_2-x_1)-k(x_2-x_3)$$
$$m\ddot x_3(t)=-k(x_3-x_2)-k(x_3-x_4)$$
$$m\ddot x_4(t)=-k(x_4-x_3)-k(x_4)$$

Simplificando las ecuaciones se llega a que: (con $\omega^2=\frac{k}{m}$)

```julia
eq1=A[1]-(w^2)*(x2-2x1)
eq2=A[2]-(w^2)*(x3-2x2+x1)
eq3=A[3]-(w^2)*(x4-2x3+x2)
eq4=A[4]-(w^2)*(-2x4+x3)
car=[eq1, eq2, eq3, eq4];
```

Vamos a obtener las matrices $\mathbb{K}$ y $\mathbb{M}$:

```julia
M=[m 0 0 0; 0 m 0 0; 0 0 m 0; 0 0 0 m]
```

```julia
k11=expand(eq1)[:coeff](x1)/(w^2)*k
k12=expand(eq1)[:coeff](x2)/(w^2)*k
k13=expand(eq1)[:coeff](x3)/(w^2)*k
k14=expand(eq1)[:coeff](x4)/(w^2)*k

k21=expand(eq2)[:coeff](x1)/(w^2)*k
k22=expand(eq2)[:coeff](x2)/(w^2)*k
k23=expand(eq2)[:coeff](x3)/(w^2)*k
k24=expand(eq2)[:coeff](x4)/(w^2)*k

k31=expand(eq3)[:coeff](x1)/(w^2)*k
k32=expand(eq3)[:coeff](x2)/(w^2)*k
k33=expand(eq3)[:coeff](x3)/(w^2)*k
k34=expand(eq3)[:coeff](x4)/(w^2)*k

k41=expand(eq4)[:coeff](x1)/(w^2)*k
k42=expand(eq4)[:coeff](x2)/(w^2)*k
k43=expand(eq4)[:coeff](x3)/(w^2)*k
k44=expand(eq4)[:coeff](x4)/(w^2)*k

k1=[k11,k12,k13,k14]
k2=[k21,k22,k23,k24]
k3=[k31,k32,k33,k34]
k4=[k41,k42,k43,k44]
K=[k1 k2 k3 k4]
```

####$(ii)$
Para calcular la posición de los carritos en función del tiempo, nada más hay que remplazar el tiempo en cada modo normal y el vector resultante va a dar las entradas como vectores cuya $i-esima$ componente representa la posición del $i-esimo$ carrito.

```julia
modo1(ω,ti,xi1)=(subs(ϕ1,t,ti)|>subs(w,ω)|>subs(ξ1,xi1))
modo2(ω,ti,xi2)=(subs(ϕ2,t,ti)|>subs(w,ω)|>subs(ξ2,xi2))
modo3(ω,ti,xi3)=(subs(ϕ3,t,ti)|>subs(w,ω)|>subs(ξ3,xi3))
modo4(ω,ti,xi4)=(subs(ϕ4,t,ti)|>subs(w,ω)|>subs(ξ4,xi4));
```

```julia
function conjuntadora(modo::Integer,ω,t0,tf,xi,h=1e-1)
    T=[t0:h:tf]
    X1=Float64[]
    X2=Float64[]
    X3=Float64[]
    X4=Float64[]
    if modo==1
        for s in T
            x=float(modo1(ω,s,xi))
            push!(X1,x[1])
            push!(X2,x[2])
            push!(X3,x[3])
            push!(X4,x[4])
        end
        
    end
    if modo==2
        for s in T
            x=float(modo2(ω,s,xi))
            push!(X1,x[1])
            push!(X2,x[2])
            push!(X3,x[3])
            push!(X4,x[4])
        end
    end
    if modo==3
        for s in T
            x=float(modo3(ω,s,xi))
            push!(X1,x[1])
            push!(X2,x[2])
            push!(X3,x[3])
            push!(X4,x[4])
        end
    end
    if modo==4
        for s in T
            x=float(modo4(ω,s,xi))
            push!(X1,x[1])
            push!(X2,x[2])
            push!(X3,x[3])
            push!(X4,x[4])
        end
    end
    return T,X1,X2,X3,X4
end
```

```julia
function plotter2(modo::Integer,ω,t0,tf,xi,h=1e-1)
    T,X1,X2,X3,X4=conjuntadora(modo,ω,t0,tf,xi,h)
    
    subplots_adjust(hspace=0.1) # separación entre las gráficas

    subplot(411)
    ax1 = gca()
    setp(ax1[:get_xticklabels](),visible=false) # No mostrar valores del eje x
    ylabel("x1(t)")
    plot(T,X1,"r-")
    axis([T[1],T[end],-2,2])
    title("Modo normal $modo")

    subplot(412, sharex=ax1)
    ax2 = gca()
    setp(ax2[:get_xticklabels](),visible=false) # No mostrar valores del eje x
    ylabel("x2(t)")
    plot(T,X2,"b-")
    axis([T[1],T[end],-2,2])
    
    subplot(413, sharex=ax2)
    ax3 = gca()
    setp(ax2[:get_xticklabels](),visible=false) # No mostrar valores del eje x
    ylabel("x3(t)")
    plot(T,X3,"y-")
    axis([T[1],T[end],-2,2])

    subplot(414, sharex=ax3)
    ylabel("$x4(t)")
    xlabel("t")
    plot(T,X4,"g-")
    axis([T[1],T[end],-2,2])
    
    
end
```

```julia
# Configuración del lienzo
fig = figure()
ax = plt.axes(xlim=(-3, 3), ylim=(-10, 10))

# Cada carrito tiene un color diferente
# b: blue, r: red
# o es para dibujar puntos en lugar de rayas
global line1 = ax[:plot]([], [], "bo", lw=2)[1]
global line2 = ax[:plot]([], [], "ro", lw=2)[1]
global line3 = ax[:plot]([], [], "yo", lw=2)[1]
global line4 = ax[:plot]([], [], "go", lw=2)[1]

# Inicialización
function init()
    global line1
    global line2
    global line3
    global line4

    line1[:set_data]([], [])
    line2[:set_data]([], [])
    line3[:set_data]([], [])
    line4[:set_data]([], [])

    return ([line1, line2, line3, line4],None)
end

# carrito 1, modo 1
y1_1 = 7.5.*ones(100)
# carrito 1, modo 2
y1_2 = 2.5.*ones(100)
# carrito 1, modo 3
y1_3 = -2.5.*ones(100)
# carrito 1, modo 4
y1_4 = -7.5.*ones(100)




# Esta función dibuja lo que aparece en cada cuadro de la animación
# i es el número de cuadro o el n-ésimo paso de tiempo
# Por ejemplo, puede ser el número de paso en el método de Euler
# La función es llamada secuencialmente por anim.FuncAnimation para cada cuadro
function animate(i)
    global line1
    global line2
    global line3
    global line4
    # i+1 para que i=0 sea x[1]
    line1[:set_data]([x1_1[i+1],x1_2[i+1],x1_3[i+1],x1_4[i+1]],[y1_1[i+1],y1_2[i+1],y1_3[i+1],y1_4[i+1]])
    line2[:set_data]([x2_1[i+1],x2_2[i+1],x2_3[i+1],x2_4[i+1]],[y1_1[i+1],y1_2[i+1],y1_3[i+1],y1_4[i+1]])
    line3[:set_data]([x3_1[i+1],x3_2[i+1],x3_3[i+1],x3_4[i+1]],[y1_1[i+1],y1_2[i+1],y1_3[i+1],y1_4[i+1]])
    line4[:set_data]([x4_1[i+1],x4_2[i+1],x4_3[i+1],x4_4[i+1]],[y1_1[i+1],y1_2[i+1],y1_3[i+1],y1_4[i+1]])
    return ([line1, line2, line3, line4],None)
end
```
