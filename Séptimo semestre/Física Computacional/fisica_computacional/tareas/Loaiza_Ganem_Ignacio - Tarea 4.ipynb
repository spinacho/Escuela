#[1] - Tom y Jerry.

Primero vamos a utilizar el método de Euler-Cromer implementado en clase: (este es específico a este problema ya que le implementamos una impresión de Tom y Jerry, sin embargo es el mismo Euler-Cromer que se utiliza en todos los problemas).

```julia
using PyPlot;
```

```julia
# Convención:
# Q[1] = x
# Q[2] = y
# V[1] = vx
# V[2] = vy
# Le añadimos una variante a las funciones para incluir la función de velocidad y posición del ratón en ambos casos.
function euler_cromer(q0,v0,t0,tf,h,f_q,f_v,X_r,V_r,paro,ϵ=1e-5)
    q = Vector{Float64}[]
    v = Vector{Float64}[]
    t = Float64[]
    vini=round(sqrt(v0[1]^2+v0[2]^2),2)
    for ti in t0:h:tf
        r_raton=X_r(ti)
        v_raton=V_r(ti)
        v_nueva = v0 + h.*f_v(ti,q0,v0,r_raton,v_raton)
        q_nueva = q0 + h.*f_q(ti,q0,v_nueva,r_raton,v_raton)
        q0 = q_nueva
        v0 = v_nueva
        push!(t,ti)
        push!(q,[q0])
        push!(v,[v0])
        if paro(q0,r_raton,ϵ) == true
            tmax=last(t)
            println("Tom atrapó a Jerry al cabo de $tmax segundos con una velocidad adimensional de $vini")
            break
        end
    end
    t,hcat(q...)',hcat(v...)'
end;
```

$$\vec{v_g}=\frac{\vec{r_r}-\vec{r_g}}{|\vec{r_r}-\vec{r_g}|}v_g$$

Dónde $v_g$ es una constante (la norma), y también lo es $v_r$ (la velocidad del ratón, aparece más adelante).

La velocidad del gato va a poder ser modelada entonces a partir de la siguiente función: (notación: $X$ es un vector, $x$ y $y$ son sus conmponentes). 
$$\vec{v_g}=f(t,X_{gato},v_g)=v_g*\frac{(x_r-x_g)\hat x+(y_r-y_g)\hat y}{\sqrt{(x_r-x_g)^2+(y_r-y_g)^2}}$$

Para cada sección vamos a utilizar entonces una posición distinta del ratón, la cual vamos a poner como una función explicita del tiempo (es facil sacar la posición del ratón en función del tiempo ya que se conoce su velocidad y su trayectoria en ambos casos).

Con condición inicial $X_{gato}(t=0)=(x_0,y_0)$. Ahora vamos a encontrar la aceleración del gato para poder aplicar Euler-Cromer. (Notación: $Dx=x_r-x_g$, $Dy=y_r-y_g$, $Ddx=v_{rx}-v_{gx}$, $Ddy=v_{ry}-v_{gy}$ y $R=\sqrt{Dx^2+Dy^2}$.
<p> Se llega a que, sea $\vec {a_g}=\vec g(t,X_{gato},V_{gato},v_g)$:
<p> 

$$\vec g(t,X_{gato},V_{gato},v_g)=\frac{v_g}{R^3}\Big(\big(Dy*(DyDdx-DxDdy)\big)\hat x + (\big(Dx*(DxDdy-DyDdx)\big)\hat y\Big)$$

Para simplificar la notación vamos a definir algunas de estas funciones:

```julia
Ddx(V_r,V_g)=V_r[1]-V_g[1]
Ddy(V_r,V_g)=V_r[2]-V_g[2]
Dx(X_r,Q_g)=X_r[1]-Q_g[1]
Dy(X_r,Q_g)=X_r[2]-Q_g[2]
R(X_r,Q_g)=sqrt(Dx(X_r,Q_g)^2+Dy(X_r,Q_g)^2)
Dddifx(X_r,V_r,Q_g,V_g)=(Dy(X_r,Q_g)*Ddx(V_r,V_g)-Dx(X_r,Q_g)*Ddy(V_r,V_g))
Dddify(X_r,V_r,Q_g,V_g)=(Dx(X_r,Q_g)*Ddy(V_r,V_g)-Dy(X_r,Q_g)*Ddx(V_r,V_g))
rap_g(V_g)=sqrt(V_g[1]^2+V_g[2]^2)
function paro_Tom(Q_g,X_r,ϵ)
    R(X_r,Q_g)<ϵ
end;
```

Ahora solo nos queda encontrar las funciones que describen la velocidad y la aceleración del gato, así como la función posición del ratón.

Vamos a implementar estas funciones:

```julia
acel_x(t,q0,vg,r_raton,v_raton)=(rap_g(vg)/(R(r_raton,q0)^3))*(Dy(r_raton,q0)*Dddifx(r_raton,v_raton,q0,vg))
acel_y(t,q0,vg,r_raton,v_raton)=(rap_g(vg)/(R(r_raton,q0)^3))*(Dx(r_raton,q0)*Dddify(r_raton,v_raton,q0,vg))
f(t,q0,vg,r_raton,v_raton)=[vg[1],vg[2]]
g(t,q0,vg,r_raton,v_raton)=[acel_x(t,q0,vg,r_raton,v_raton),acel_y(t,q0,vg,r_raton,v_raton)];
```

###(a)
Primero vamos a definir la función que describe la posición del ratón:
$$\vec{v_r}=v_r\hat x$$
$$\vec{X_r}=v_rt\hat x$$
Implementemos la función:

```julia
v_r_a(t)=[1,0]
X_r_a(t)=[t,0];
```

(Vamos a tomar la velocidad del ratón como 1, de forma que la velocidad del gato va a ser el cociente de velocidades. $v_g=1$ es equivalente entonces a que $v_g=v_r$).

```julia
function graficadora_Tom_a(q0,vini,tf=30)
    r0=[X_r_a(0)[1]-q0[1],X_r_a(0)[2]-q0[2]]
    norm=sqrt(r0[1]^2+r0[2]^2)
    rnorm=(1/norm).*r0
    v0=(vini).*[rnorm]
    t0=0
    h=0.0001
    f_q=f
    f_v=g
    X_r=X_r_a
    V_r=v_r_a
    t,M,Mdot=euler_cromer(q0,v0,t0,tf,h,f_q,f_v,X_r,V_r,paro_Tom,1e-3)
    equis=M[:,1]
    ye=M[:,2]
    plot(equis,ye, label="Velocidad adimensional de $vini ")
    xlabel(L"$x$")
    ylabel(L"$y$")
    title("Persecución de Tom y Jerry")
    legend(loc=0, fontsize=14)
end;
```

Podemos notar cómo sólo lo atrapó para las velocidades estrictamente mayores a la de Jerry (ya que se tomó, sin pérdida de generalidad, $v_{Jerry}=1$ de forma adimensional). De esta forma, obtuvimos que Tom atrapa a Jerry cuando $v_g>v_r$. Sólo se hizo la prueba para la posición inicial del gato Tom en $(-1,-1)$.

#[2] - Una carrera singular
Del enunciado de la tarea:
Definiendo las variables adimensionales,

$$
\rho = r/d, \quad \tau = t/t_0 \;\text{con}\; t_0 = \sqrt{d/g}, \quad \mu=m_1/m_2, 
$$

las ecuaciones de movimiento son:

$$
\begin{array}{l}
\rho'' = \frac{1}{\mu+1}(\rho\theta'^2 + \cos\theta)\\
\theta'' = -\frac{1}{\rho}(2\rho'\theta' + \sin\theta).
\end{array}
$$

La energía total del sistema en términos de variables adimensionales es:

$$
E/E_0 = (K_1 + K_2 + U_2)/E_0, \quad E_0 = m_2gd
$$

con ($\nu = \rho', \omega = \theta'$)

$$
\begin{array}{l}
K_1/E_0 = \frac12 \mu\nu^2\\
K_2/E_0 = \frac12[\nu^2 + (\rho\omega)^2]\\
U_2/E_0 = -\rho\cos\theta
\end{array}
$$

```julia
# Convención:
# Q[1] = ρ
# Q[2] = θ
# V[1] = ν
# V[2] = ω
function euler_cromer2(q0,v0,t0,tf,h,f_q,f_v,paro,μ,ϵ=1e-5)
    q = Vector{Float64}[]
    v = Vector{Float64}[]
    t = Float64[]
    for ti in t0:h:tf
        v_nueva = v0 + h.*f_v(ti,q0,v0,μ)
        q_nueva = q0 + h.*f_q(ti,q0,v_nueva,μ)
        q0 = q_nueva
        v0 = v_nueva
        push!(t,ti)
        push!(q,[q0])
        push!(v,[v0])
        if paro(q0,ϵ) == 3
            println("El cuadro 1 llegó al borde al mismo tiempo que el cuadro 2 chocó con la pared al cabo de $ti (tiempo adimensional) para μ=$μ")
            break
        end
            if paro(q0,ϵ) == 1
            println("El cuadro 1 llegó al borde al cabo de $ti (tiempo adimensional) para μ=$μ")
            break
        end
             if paro(q0,ϵ) == 2
            println("El cuadro 2 chocó con la pared al cabo de $ti (tiempo adimensional) para μ=$μ")
            break
        end
    end
    t,hcat(q...)',hcat(v...)'
end;
```

Ahora vamos a implementar las funciones para el método:

```julia
fq2(t,Q,V,μ)=[V[1],V[2]]
fv2_ρ(t,Q,V,μ)=(1/(μ+1))*(Q[1]*(V[2]^2)+cos(Q[2]))
fv2_θ(t,Q,V,μ)=(-1/Q[1])*(2*V[1]*V[2]+sin(Q[2]))
fv2(t,Q,V,μ)=[fv2_ρ(t,Q,V,μ),fv2_θ(t,Q,V,μ)]
K1(ρ,θ,ν,ω,μ)=0.5*μ*ν
K2(ρ,θ,ν,ω,μ)=0.5*(ν^2+((ρ*ω)^2))
U2(ρ,θ,ν,ω,μ)=-ρ*cos(θ);
```

Finalmente, vamos a implementar la función de paro (queremos parar si $\theta=0$ ó si $r=2d$, sea si $\rho=2$).

```julia
function paro_carrera(Q,ϵ)
    if Q[1]>2 && Q[2]<0
        return 3
    end
    if Q[1]>2
        return 1
    end
    if Q[2]<0
        return 2
    end
    0
end;
```

Hay que recordar que $x(\tau)=r(\tau)-2d=d(\rho(\tau)-2)$

```julia
function graficadora_carrera(q0,v0,μ,d)
    t0=0
    tf=10
    h=0.0001
    f_q=fq2
    f_v=fv2
    t,M,Mdot=euler_cromer2(q0,v0,t0,tf,h,f_q,f_v,paro_carrera,μ,1e-3)
    rho=M[:,1]
    theta=M[:,2]
    equis=[d*(rho[i]-2) for i in 1:length(rho)]

    subplots_adjust(hspace=0.01) # separación entre las gráficas

    subplot(311)
    ax1 = gca()
    setp(ax1[:get_xticklabels](),visible=false) # No mostrar valores del eje x
    ylabel(L"$\rho(t)$")
    plot(t,rho,"r-")
    axis([0,last(t),first(rho),last(rho)])

    subplot(312, sharex=ax1)
    ax2 = gca()
    setp(ax2[:get_xticklabels](),visible=false) # No mostrar valores del eje x
    ylabel(L"$\theta(t)$")
    xlabel(L"$t$")
    plot(t,theta)
    axis([0,last(t),last(theta),first(theta)])

    subplot(313, sharex=ax2)
    ylabel(L"$x(t)$")
    xlabel(L"$t$")
        plot(t,equis,"g-")
    axis([0,last(t),last(equis),first(equis)])
    
end;
```

###(2)
Vamos a introducir la función de la posición del centro de masa:
$$\frac{x_{\mathrm{cm}}}{d} = \frac{\rho(\mu + \sin\theta) -2\mu}{\mu + 1}.$$

```julia
function graficadora_centro_de_masa(q0,v0,μ,d)
    t0=0
    tf=10
    h=0.0001
    f_q=fq2
    f_v=fv2
    t,M,Mdot=euler_cromer2(q0,v0,t0,tf,h,f_q,f_v,paro_carrera,μ,1e-3)
    rho=M[:,1]
    theta=M[:,2]
    x_cm=[d*(rho[i]*(μ+sin(theta[i]))-2μ)/(μ+1) for i in 1:length(rho)]
    plot(t,x_cm, label=L"x_{cm}(\tau)")
    xlabel(L"Tiempo adimensional $\tau$ ($s$)")
    ylabel(L"Posicion del centro de masa $x_{cm}$ ($m$)")
    title("Movimiento del centro de masa")
    legend(loc=0, fontsize=14)
end;
```

```julia
graficadora_centro_de_masa([1.0,π/2],[0,0],1,2);
```

Podemos ver como al principio el centro de masa práctimamente no se mueve, y según va avanzando el tiempo se empieza a mover mucho más. Esto se debe a que, al principio, como el ángulo $\theta$ es muy pequeño, la fuerza que se aplica en el primer bloque (la tensión de la cuerda, la cual siempre apunta en la dirección $\hat x$ cuando se aplica en este bloque) es muy pequeña, al igual que la componente de la fuerza en $\hat x$ en el segundo bloque, de forma que los movimientos de ambos bloques se compenzan, teniendo que la componente en $\hat x$ del centro de masa se queda inmóvil. Sin embargo, cuando $\theta$ pasa los $\frac{\pi}{4}$ radianes, se tiene que la componente de la fuerza en $\hat x$ del bloque 2 llega a un máximo y comienza a disminuir, mientras que la fuerza en el bloque 1 sigue aumentando, teniendo que el movimiento ya no se compenza en el centro de masa, por lo cual este se comienza a mover a la derecha. 

###(3)
Ya implementamos las funciones de las energías, sólo hay que calcularlas y graficarlas:

```julia
function graficadora_energias(q0,v0,μ,d)
    t0=0
    tf=10
    h=0.0001
    f_q=fq2
    f_v=fv2
    t,M,Mdot=euler_cromer2(q0,v0,t0,tf,h,f_q,f_v,paro_carrera,μ,1e-3)
    rho=M[:,1]
    theta=M[:,2]
    nu=Mdot[:,1]
    omega=Mdot[:,2]
    ll=length(rho)
    ka1=[K1(rho[i],theta[i],nu[i],omega[i],μ) for i in 1:ll]
    ka2=[K2(rho[i],theta[i],nu[i],omega[i],μ) for i in 1:ll]
    u2=[U2(rho[i],theta[i],nu[i],omega[i],μ) for i in 1:ll]
    E=ka1+ka2+u2
    plot(t,ka1, label=L"\frac{K_1}{E_0}")
    plot(t,ka2, label=L"\frac{K_2}{E_0}")
    plot(t,u2, label=L"\frac{U2}{E_0}")
    plot(t,E, label=L"\frac{E}{E_0}")
    xlabel(L"Tiempo adimensional $\tau$ ($s$)")
    ylabel("Energía adimensional")
    title("Energías del sistema")
    legend(loc=0, fontsize=14)
end;
```

```julia
graficadora_energias([1.0,π/2],[0,0],1,2);
```

Podemos ver como la energía total se conserva aproximadamente.

###(5)
Ahora vamos a hacer una función que nos grafique el tiempo en función de $\mu$.
<p> La $\mu$ que nos va a interesar va a ser entonces la correspondiente al pico en la función del tiempo, ya que claramente el tiempo va a hacer una curva primero creciente y luego decreciente en función de $\mu$, (ya que, al principio, cuando el bloque 1 llega más rápido a la polea el método se va haciendo más y más lento, hasta que llega al punto donde coinciden (el pico), y a partir de este punto el bloque 2 va llegando más y más rápido a la pared, disminuyendo el tiempo en función de la $\mu$).
<p> La impresión de los valores por la función nos va a ayudar también para encontrar esta $\mu$ (la podríamos encontrar sólo evaluando, sin necesidad de graficar ningún valor).

```julia
function maximos(q0,v0,μ,d)
    t,M,Mdot=euler_cromer2(q0,v0,0,3,0.001,fq2,fv2,paro_carrera,μ,1e-3)
    rho=M[:,1]
    theta=M[:,2]
    last(t)
end;
```

```julia
MU=[1:0.01:2]
TMAX=[maximos([1.0,π/2],[0,0],μ,2) for μ in MU]
plot(MU,TMAX, label=L"Tiempo $\tau$")
xlabel(L"$\mu$")
ylabel(L"Tiempo adimensional $\tau$")
title(L"Tiempos del sistema en funcion de la $\mu$")
legend(loc=0, fontsize=14);
```

Podemos ver como el valor para el cual se obtiene un empate es $\mu\approx 1.42$.
<p> Vamos a hacer una partición de $\mu$ entre $1.41$ y $1.43$ para obtener con mayor exactitud el valor de $\mu$.

```julia
MU=[1.41:0.0001:1.43]
TMAX=[maximos([1.0,π/2],[0,0],μ,2) for μ in MU]
plot(MU,TMAX, label=L"Tiempo $\tau$")
xlabel(L"$\mu$")
ylabel(L"Tiempo adimensional $\tau$")
title(L"Tiempos del sistema en funcion de la $\mu$")
legend(loc=0, fontsize=14);
```

Tomemos la $\mu$ a la mitad del pico, sea $\mu=1.421$
