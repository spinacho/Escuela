####(1) - Escribe una función recursiva para calcular una raíz de una función $f$ en un intervalo $[a,b]$ con el método de bisección. Verifica que funcione al calcular raíces de distintas funciones.

```julia
function raiz(f,x,a=0,b=f(x)+1,i=0)
    ϵ=1e-16
    c=0.5*(a+b)
    if c-a<ϵ || i>100
        return c
    else
        i=i+1
        if c^2>f(x)
            b=c
            raiz(f,x,a,b,i)
        else
            a=c
            raiz(f,x,a,b,i)
        end
    end
end
```

Para esta función, pusimos la variable $i$ cómo un contador de iteraciones, de forma que si al cabo de la iteración número 100 no ha terminado el programa (sea que el intervalo es aún mayor que la tolerancia), lo termina a la fuerza, arrojando la aproximación en la que va. Se utilizó una tolerancia de $\epsilon=10^{-16}$.

```julia
raiz(x->x^2,2)
```

```julia
raiz(x->x,2)
```

```julia
sqrt(2)
```

####(2) - Considera la colisión en una dimensión de una partícula haciendo movimiento armónico simple con frecuencia angular $\omega$ y una partícula libre con velocidad $v$.
####(a) Escribe una función que toma los parámetros del oscilador y de la partícula y grafica sus respectivas posiciones en el tiempo, suponiendo que no hay colisión.

Tomaremos, sin pérdida de generalidad, al oscilador centrado en $x=0$, con posición inicial $x_0$ y velocidad inicial $v_0$, y a la partícula la tomaremos en la posición inicial $x_1$, con velocidad $-v$ (ya que se va acercando al oscilador). Dentro del programa vamos a definir las constantes $A$ y $B$ de forma que, sin importar los valores iniciales, la función describa el movimiento de forma correcta. Para hacer esto hay que resolver la ecuación diferencial:
$$\ddot{x}=-\omega^2 x$$
<p>Con condiciones iniciales $\omega$, $x_0$ y $v_0$.
<p>Tomando la solución de la forma:
$$x(t)=A\cos(\omega t) +B\sin(\omega t)$$
Por lo cual:
$$\dot{x}(t)=-\omega A\sin(\omega t) +\omega B\cos(\omega t)$$
Evaluando en $0$, llegamos a que:
$$x_0=A, \ \ \ v_0=\omega B$$
De donde:
$$A=x_0, \ \ \ B=\frac{v_0}{\omega}$$

```julia
function oscilador(t,ω,x0=1,v0=0)
    A=x0
    B=v0/ω
    A*cos(ω*t)+B*sin(ω*t)
end
```

```julia
function particula(t,x1=10,v=5)
    x1-v*t
end
```

```julia
using PyPlot
```

```julia
function graficadora(T,ω,x0=1,v0=0,x1=10,v=5,paso=0.01)
    TIEMPO=[0:paso:T]
    X_OSC=[oscilador(t,ω,x0,v0) for t in TIEMPO]
    X_PART=[particula(t,x1,v) for t in TIEMPO]
    plot(TIEMPO , X_PART, label="Partícula")
    plot(TIEMPO , X_OSC, label="Oscilador")
    xlabel(L"Tiempo $t$")
    ylabel(string("Posición ",L"$x$"))
    title("Movimiento de la partícula y del oscilador en función del tiempo")
    legend(loc=0, fontsize=14)
end
```

```julia
graficadora(10,1)
```

La función "graficadora" toma entonces los aprámetros del oscilador y de la partícula y grafica sus respectivas posiciones en el tiempo.

```julia
graficadora(10,3,5,2,10,3)
```

####(b) - Dibuja (en gráficas separadas) las trayectorias de las dos partículas para unas cuantas posibilidades de los parámetros. ¿Cuántas "colisiones" puede haber? ¿Cuál es la buena?

```julia
function graficadora_particula(T,x1=10,v=5,paso=0.01)
    TIEMPO=[0:paso:T]
    X_PART=[particula(t,x1,v) for t in TIEMPO]
    plot(TIEMPO , X_PART, label="Partícula")
    xlabel(L"Tiempo $t$")
    ylabel(string("Posición ",L"$x$"))
    title("Movimiento de la partícula en función del tiempo")
    legend(loc=0, fontsize=14)
end
```

```julia
function graficadora_oscilador(T,ω,x0=1,v0=0,paso=0.01)
    TIEMPO=[0:paso:T]
    X_OSC=[oscilador(t,ω,x0,v0) for t in TIEMPO]
    plot(TIEMPO , X_OSC, label="Oscilador")
    xlabel(L"Tiempo $t$")
    ylabel(string("Posición ",L"$x$"))
    title("Movimiento del oscilador en función del tiempo")
    legend(loc=0, fontsize=14)
end
```

#### <p>(c) - Escribe una función para calcular el tiempo de colisión de las dos partículas, sin garantizar cuál es, utilizando la función que escribiste en la pregunta [1]. Y: <p>(d) - ¿Cómo puedes garantizar que se encuentre la colisión adecuada? Impleméntalo.

Para escribir esta función, vamos a crear una función cuya salida sea el conjunto de valores de $t$ en los cuales se tiene que $x_{oscilador}(t)-x_{particula}(t)=0$. Sobre este conjunto, utilizamos la función **minimum** para obtener el menor tiempo mínimo para el cual se cumple esta relación, sea la primera "colisión" (la que de verdad es una colisión). Cabe notar que, como *Julia* trabaja con puntos flotantes, tenemos que definir una tolerancia $\epsilon$ para la cual la diferencia sea más pequeña que $\epsilon$, ya que si buscamos los valores en los cuales la diferencia sea $0$ entonces *Julia* no va a encontrar ningún valor.

```julia
function conjuntadora(tolerancia,paso=1e-5,T=10,ω=1,x0=1,v0=0,x1=10,v=5)
    diferencia(t)=oscilador(t,ω,x0,v0)-particula(t,x1,v)
    TIEMPO=[0:paso:T]
    salida=Float64[]
    for t in TIEMPO
        if abs(diferencia(t)) <= tolerancia
            push!(salida,t)
        end
    end
    salida
end
```

```julia
conjuntadora(1e-5)
```

Se va a hacer una función que nos permita, además de calcular el tiempo de colisión, graficar las posiciones para poder verificar que se obtuvo el resultado deseado:

```julia
function La_Mera_Mera(tolerancia,paso=1e-5,T=10,ω=1,x0=1,v0=0,x1=10,v=5)
    tiempo=minimum(conjuntadora(tolerancia,paso,T,ω,x0,v0,x1,v))
    pos=round(particula(tiempo,x1,v),4)
    println("El tiempo de colisión es t=$tiempo, y se efectúa en la posición x=$pos")
    graficadora(T,ω,x0,v0,x1,v,paso)
end
```

```julia
La_Mera_Mera(0.01, 0.0001, 10, 7, 5, 0, 10, 1)
```

Finalmente, vamos a hacer una manipulación interactiva que nos permita obtener los tiempos cuando vamos variando los valores iniciales:

```julia
@manipulate for T in 5:5:20, ω in 0:0.1:10, x0 in 0:0.1:10, v0 in 0:0.1:10, x1 in 0:0.1:30, v in 0:0.1:10
    minimum(conjuntadora(0.01, 0.0001, T,ω,x0,v0,x1,v))
end
```

####(2) - Implementa el método de Newton para calcular una raiz de una función $f$, dadas una adivinanza inicial $x_0$ y la derivada $f'$ (que por el momento se implementa a mano en otra función).

####(a) - Verifica que tu función sirve para encontrar raices de la función $f(x) = x^2 - 2$. 

Primero vamos a implementar una función derivada:

```julia
function derivada(f,x, h=1,i=0)
    ϵ=1e-10
    x=big(x)
    der(h)=(f(x+h)-f(x-h))/2h
    while abs(der(h)-der(10h))>ϵ
        h=h/10
        i=i+1
    end
    der(h)
end
```

Para definir el método de Newton-Raphson, tenemos que la iteración:
$$x_{n+1}=x_n-\frac{f(x_n)}{f'(x_n)}$$
De aquí construimos a la función Newton, a la cual le metemos una función f, una adivinanza inicial x, y calcula el valor *cercano* para el cual $f(x_0)=0$ (con una tolerancia $\epsilon=10^{-16}$. Se le puso un contador de iteraciones a la función para que, si al cabo de 100 iteraciones no encuentra un valor apropiado, la función regresa la iteración $x_{100}$.

```julia
function Newton(f,x,i=0) #ve en que valor la función f se anula
    ϵ=1e-16
    x=big(x)
    while abs(f(x))>ϵ && i<100
        x=x-(f(x)/derivada(f,x))
        i=i+1
    end
    x=round(float64(x),16)
end
```

```julia
Newton(x-> x^2-2,2)
```

```julia
Newton(x-> x^2-2,-2)
```

```julia
sqrt(2)
```

Notamos que, para los valores iniciales de $2$ y $-2$, se obtuvieron excelentes aproximaciones de la raíz de $2$.

####(b) - Intenta con distintos valores iniciales. Dibuja una gráfica que muestra a cuál raiz llega el método para distintos valores iniciales.


```julia
VALORES=[Newton(x-> x^2-2,ex) for ex in [-100:0.1:100]]
```

```julia
plot(VALORES)
```

Podemos ver como, para valores iniciales en $[-100,0)$ el método nos regresa su aproximación (bastante buena) de $-\sqrt2$, para valores iniciales en $(0,100]$ nos regresa la de $\sqrt2$
